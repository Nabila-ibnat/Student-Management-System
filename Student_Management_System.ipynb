```python
# modules/student.py

class Student:
    def __init__(self, name, age, marks):
        self.name = name
        self.age = age
        self.marks = marks
```

```python
# modules/grades.py

def calculate_grade(mark):
    mark = float(mark)
    if mark >= 80:
        return "A+"
    elif mark >= 70:
        return "A"
    else:
        return "B"
```

```python
# modules/utils.py

def print_line(char="-", times=50):
    print(char * times)

def total_marks(*marks):
    return sum(marks)
```

```python
# modules/operations.py

import os

DATA_FILE = "students.txt"

def load_students():
    students = []
    if not os.path.exists(DATA_FILE):
        return students
    with open(DATA_FILE, "r") as f:
        for line in f:
            line = line.strip()
            if line:
                parts = line.split(",")
                if len(parts) == 3:
                    students.append(Student(parts[0], parts[1], parts[2]))
    return students

def save_students(students):
    with open(DATA_FILE, "w") as f:
        for s in students:
            f.write(f"{s.name},{s.age},{s.marks}\n")

def add_student():
    print_line("=")
    print("ADD STUDENT".center(50))
    print_line("=")
    name  = input("Enter Student Name  : ")
    age   = input("Enter Student Age   : ")
    marks = input("Enter Student Marks : ")
    students = load_students()
    students.append(Student(name, age, marks))
    save_students(students)
    print("Student Added Successfully!")
    print_line("=")

def view_students():
    print_line("=")
    print("STUDENT LIST".center(50))
    print_line("=")
    students = load_students()
    if not students:
        print("No students found.")
        print_line("=")
        return
    print(f"{'Name':<20} {'Age':<6} {'Marks':<8}")
    print_line()
    for s in students:
        print(f"{s.name:<20} {s.age:<6} {s.marks:<8}")
    print_line("=")
    print(f"Total Students: {len(students)}")
    print_line("=")

def search_student():
    print_line("=")
    print("SEARCH STUDENT".center(50))
    print_line("=")
    name = input("Enter Student Name: ").lower()
    students = load_students()
    found = [s for s in students if name in s.name.lower()]
    if not found:
        print("Student Not Found!")
    else:
        for s in found:
            print("Student Found!")
            print_line()
            print(f"Name  : {s.name}")
            print(f"Age   : {s.age}")
            print(f"Marks : {s.marks}")
    print_line("=")

def update_student():
    print_line("=")
    print("UPDATE STUDENT".center(50))
    print_line("=")
    name = input("Enter Student Name: ").lower()
    students = load_students()
    for i, s in enumerate(students):
        if s.name.lower() == name:
            print("Current Information")
            print_line()
            print(f"Name  : {s.name}")
            print(f"Age   : {s.age}")
            print(f"Marks : {s.marks}")
            new_name  = input("Enter New Name  : ")
            new_age   = input("Enter New Age   : ")
            new_marks = input("Enter New Marks : ")
            students[i] = Student(new_name, new_age, new_marks)
            save_students(students)
            print("Student Updated Successfully!")
            print_line("=")
            return
    print("Student Not Found!")
    print_line("=")

def delete_student():
    print_line("=")
    print("DELETE STUDENT".center(50))
    print_line("=")
    name = input("Enter Student Name: ").lower()
    students = load_students()
    new_list = [s for s in students if s.name.lower() != name]
    if len(new_list) == len(students):
        print("Student Not Found!")
    else:
        save_students(new_list)
        print("Student Deleted Successfully!")
    print_line("=")
```

```python
# main.py

def main():
    while True:
        print_line("=")
        print("STUDENT MANAGEMENT SYSTEM".center(50))
        print_line("=")
        print("1. Add Student")
        print("2. View Students")
        print("3. Search Student")
        print("4. Update Student")
        print("5. Delete Student")
        print("6. Exit")
        print_line("=")
        choice = input("Enter your choice: ").strip()

        if choice == "1":
            add_student()
        elif choice == "2":
            view_students()
        elif choice == "3":
            search_student()
        elif choice == "4":
            update_student()
        elif choice == "5":
            delete_student()
        elif choice == "6":
            print_line("=")
            print("Thank You For Using".center(50))
            print("STUDENT MANAGEMENT SYSTEM".center(50))
            print("Program Closed Successfully!".center(50))
            print_line("=")
            break
        else:
            print("Invalid choice. Please enter 1-6.")

main()
```
